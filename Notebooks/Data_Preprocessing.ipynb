<h2>DATA PREPROCESSING</h2>

<h3>IDS_ALLCountries_Data</h3>

```python
import pandas as pd
import numpy as np
```

```python
df_IDS_ALLCountries_Data = pd.read_csv('IDS_ALLCountries_Data.csv', encoding='latin1')
```

```python
df_IDS_ALLCountries_Data.isnull().sum()
```

<p>
An analysis of missing values revealed that the columns representing the years 2025 to 2032 contain over 50,000 null records each. Since these columns have insufficient data and are unlikely to contribute meaningful insights to the analysis, they were excluded from the dataset to improve data quality and reduce unnecessary complexity.
<p>


```python
df_IDS_ALLCountries_Data = df_IDS_ALLCountries_Data.drop(columns = ['2025','2026','2027','2028','2029','2030','2031','2032'])
```

```python
df_IDS_ALLCountries_Data.isnull().sum()
```

```python
df_IDS_ALLCountries_Data[df_IDS_ALLCountries_Data.duplicated()]

```

<h5> Two duplicate records were identified in the dataset. As all fields in these records contained null values and did not contribute any meaningful information, they were removed during the data cleaning process.</h5>

```python
df_IDS_ALLCountries_Data.drop([62979,62980], inplace = True)
```

```python
df_IDS_ALLCountries_Data[df_IDS_ALLCountries_Data.duplicated()]
```

```python
year_cols = [str(year) for year in range(2000, 2025)]

df_IDS_ALLCountries_Data = pd.melt(
    df_IDS_ALLCountries_Data,
    id_vars=[col for col in df_IDS_ALLCountries_Data.columns if col not in year_cols],
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)
```

```python
df_IDS_ALLCountries_Data.columns
```

```python
df_IDS_ALLCountries_Data
```

```python
df_IDS_ALLCountries_Data.columns
```

```python
df_IDS_ALLCountries_Data.info()
```

```python
df_IDS_ALLCountries_Data = df_IDS_ALLCountries_Data.dropna(subset=['Value'])
```

```python
df_IDS_ALLCountries_Data.nunique()
```

```python
cols_to_drop = df_IDS_ALLCountries_Data.columns[
    df_IDS_ALLCountries_Data.nunique() == 1
]

df_IDS_ALLCountries_Data=df_IDS_ALLCountries_Data.drop(columns=cols_to_drop)
```

```python
df_IDS_ALLCountries_Data
```

<p>
The columns <b>Counterpart-Area Name</b> and <b>Counterpart-Area Code</b> each contain only a single unique value across all records. As these columns do not provide any variability or contribute meaningful information for analysis, they were removed from the dataset to reduce redundancy and improve data efficiency.
<p>

```python
df_IDS_ALLCountries_Data.info()
```

<p>The dataset originally contained separate columns for each year from 2000 to 2024, resulting in a large number of columns. To improve data organization and make the dataset more suitable for analysis and visualization, these year-specific columns were transformed into a long format using a single Year column and a corresponding Value column.</p>

<p>In this transformation, the year information from the columns 2000–2024 was consolidated into the Year column, while the corresponding data values were stored in the Value column. This significantly reduced the number of columns in the dataset, although the number of rows increased as each year-value pair became a separate record.</p>

<p>Furthermore, the Value column represents the primary measure of interest in this dataset and is essential for meaningful analysis. Therefore, all records containing null (missing) values in the Value column were removed to ensure data quality and maintain the reliability of subsequent analysis and visualizations.</p>

```python
df_IDS_ALLCountries_Data.to_csv('Cleaned_IDS_ALLCountries_Data.csv',index = False)
```
